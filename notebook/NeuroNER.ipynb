## 1.기존 Pre-trained model에 약간의 학습을 가미한 정도만으로도 학습이 되는가?  
되는 것 같습니다.  

### 논문에 제시된 pretrained model을 이용한 training mode에 관한 내용(https://arxiv.org/pdf/1705.05487.pdf)
NeuroNER has 3 modes of operation: 중  
  training mode (from pretrained model): the dataset folder must have train and valid sets. Test and deployment sets are optional.  
  
  
  
### 그렇다면 그 방법은? = traing mode(from .pretrained model)  
학습을 시키려면 'src/parameters.ini'에서 train_model=True 로 수정하거나  
터미널에서 명령어 입력시 'python3.5 main.py --train_model=False'과 같은 옵션을 주면됩니다.


github에 나온 예시 명령어를 보면 :  
python3.5 main.py --train_model=False --use_pretrained_model=True --dataset_text_folder=../data/example_unannotated_texts --pretrained_model_folder=../trained_models/conll_2003_en  


이는 'conll_2003_en이라는 pretrained_model을 사용해서 unannotated_texts 파일을 annotate 해라' 라는 뜻 같습니다.  





여기서 '--train_model=False'을 '--train_model=True' 로 바꿔보면 아래와 같은 에러가 나타납니다.  
![title](img/train_model_true_error.png)

이 것으로 보아
--train_model=True에서 dataset은 train 할 dataset을 의미하고  
--train_model=False에서 dataset은 annotate할 dataset을 의미하는 것 같습니다.  

그래서 --train_model=True 시 dataset에서는 train folder와 validation folder가 필요한 것이죠.  


### 결론 
traing mode(from .pretrained model)를 사용하시려면  


python3.5 main.py --train_model=True --use_pretrained_model=True  


을 입력하시면 되며, '--dataset_text_folder= [텍스트 데이터] '에서의 텍스트 데이터는 꼭 train folder와 validation folder 가 필요합니다.

상세 설정은 'src/parameters.ini'에서 할 수 있습니다.

### 테스트 방법
- brat 실행


- 기존 pretrained model 가지고 다른 텍스트 annotate 실시  
   -> unanotated_text 내 문장을 새로운 문장으로 바꾼 뒤 brat으로 확인하거나  
   -> unanotated_text 폴더 자체를 바꿈
   
   명령어 :  
   python3.5 main.py --train_model=False --use_pretrained_model=True --dataset_text_folder=../data/example_unannotated_texts --pretrained_model_folder=../trained_models/conll_2003_en  


- 기존 pretrained model에 새로운 데이터를 더 해서 학습시킴
    -> 새로운 데이터 셋을 구하여, 명령어에 unanotated_text 경로 대신 새로운 데이터셋 경로 추가
    -> 새로운 데이터 셋은 반드시 train.txt와 validation.txt로 나눠져야 함
    
    
    명령어 :
    python3.5 main.py --train_model=False --use_pretrained_model=True --dataset_text_folder=[새로운 데이터셋 경로]  --pretrained_model_folder=../trained_models/conll_2003_en
